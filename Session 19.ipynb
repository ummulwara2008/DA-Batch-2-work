```python
for i in survey:
    years = i[0]

    if years <5:
        category = "< 5 Years"
    elif years < 10:
        category = "5-10 Years"
    elif years  < 15:
        category = "< 10-15 Years"
    elif years  < 20:
        category = "< 15-20 Years"
    else:
        category = " 25+ Years"

    i.append(category)
```

```python
lessthan5 = []
fiveto10 = []
tento15 = []
fifteento20 = []
twentyto25 = []
morethan25 = []

for i in survey:
    if i[-1] == "< 5 Years":
        lessthan5.append(i[-2])
    elif i[-1] == "5-10 Years":
        fiveto10.append(i[-2])
    elif i[-1] == "10-15 Years":
        tento15.append(i[-2])
    elif i[-1] == "15-20 Years":
        fifteento20.append(i[-2])
    elif i[-1] == "20-25 Years":
        twentyto25.append(i[-2])
    else:
        morethan25.append(i[-2])
```
