```python
df[(df['sql_user'] == 'TensorFlow') & (df['r_user'] == 'PyTorch')]
```

```python
df[(df['python_user'] == 'TensorFlow')& ~ (df['r_user'] == "PyTorch")]
```

```python
df[(df['sql_user'] == 'TensorFlow')| ~ (df['r_user'] == "PyTorch")]
```
